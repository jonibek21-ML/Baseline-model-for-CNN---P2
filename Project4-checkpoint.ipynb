# 🖼 Data Augmentation (Transforms)

Bu bosqichda training dataset uchun **data augmentation** qo‘llaniladi:

Transformlar:

- Resize → Rasmlarni bir xil o‘lchamga keltirish
- RandomResizedCrop → Random crop
- Horizontal Flip → Tasodifiy aylantirish
- Rotation → Tasodifiy burish
- ColorJitter → Yorug‘lik va rang o‘zgarishi
- RandomAffine → Scale va translate o‘zgarishlari

Maqsad:
Modelni overfittingdan saqlash va generalization oshirish

```python
from torchvision import transforms

tr_transform = transforms.Compose([
    transforms.Resize((256,256)),

    transforms.RandomResizedCrop(224, scale=(0.85,1.0)),

    transforms.RandomHorizontalFlip(p=0.5),

    transforms.RandomRotation(15),

    transforms.ColorJitter(
        brightness=0.15,
        contrast=0.15,
        saturation=0.10,
        hue=0.02
    ),

    transforms.RandomAffine(
        degrees=0,
        translate=(0.05,0.05),
        scale=(0.95,1.05)
    ),

    transforms.ToTensor(),

    transforms.Normalize(
        mean=[0.485,0.456,0.406],
        std=[0.229,0.224,0.225]
    )
])

vl_transform = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor(),

    transforms.Normalize(
        mean=[0.485,0.456,0.406],
        std=[0.229,0.224,0.225]
    )
])
```

# 📂 Dataset Split (Train / Validation / Test)

Dataset quyidagi qismlarga bo‘lindi:

- 70% → Training
- 15% → Validation
- 15% → Testing

Stratified split ishlatilgan:

Bu class distributionni barcha dataset qismlarida bir xil saqlaydi.

```python
dataset = torchvision.datasets.ImageFolder("D:/data/train")
targets = numpy.array(dataset.targets)

train_idx, temp_idx = train_test_split(
    numpy.arange(len(targets)),
    test_size=0.3,
    stratify=targets,
    random_state=42
)

val_idx, test_idx = train_test_split(
    temp_idx,
    test_size=0.5,
    stratify=targets[temp_idx],
    random_state=42
)

train_ds = Subset(
    torchvision.datasets.ImageFolder("D:/data/train", transform=tr_transform),
    train_idx
)

val_ds = Subset(
    torchvision.datasets.ImageFolder("D:/data/train", transform=vl_transform),
    val_idx
)

test_ds = Subset(
    torchvision.datasets.ImageFolder("D:/data/train", transform=vl_transform),
    test_idx
)

tr_load = torch.utils.data.DataLoader(train_ds, batch_size=16, shuffle=True)
vl_load = torch.utils.data.DataLoader(val_ds, batch_size=16)
te_load = torch.utils.data.DataLoader(test_ds, batch_size=16)

classes = dataset.classes
print(classes)
```

# 🧠 CNN Model Architecture

Bu model Sequential CNN arxitekturadan iborat:

Layerlar:

Conv2D → Feature extraction
BatchNorm → Training stability
ReLU → Activation function
MaxPool → Spatial reduction
Dropout → Overfittingni kamaytirish

Oxirida:

Fully Connected layer classification uchun ishlatiladi

```python
model = nn.Sequential(
    nn.Conv2d(3,32,3,padding=1),
    nn.BatchNorm2d(32),
    nn.ReLU(),
    nn.Conv2d(32,32,3,padding=1),
    nn.BatchNorm2d(32),
    nn.ReLU(),
    nn.MaxPool2d(2),
    nn.Dropout(0.3),

    nn.Conv2d(32,64,3,padding=1),
    nn.BatchNorm2d(64),
    nn.ReLU(),
    nn.Conv2d(64,64,3,padding=1),
    nn.BatchNorm2d(64),
    nn.ReLU(),
    nn.MaxPool2d(2),
    nn.Dropout(0.3),

    nn.Conv2d(64,128,3,padding=1),
    nn.BatchNorm2d(128),
    nn.ReLU(),
    nn.Conv2d(128,128,3,padding=1),
    nn.BatchNorm2d(128),
    nn.ReLU(),
    nn.MaxPool2d(2),
    nn.Dropout(0.3),
    
    nn.Conv2d(128,256,3,padding=1),
    nn.BatchNorm2d(256),
    nn.ReLU(),
    nn.Conv2d(256,256,3,padding=1),
    nn.BatchNorm2d(256),
    nn.ReLU(),
    nn.MaxPool2d(2),
    nn.Dropout(0.3),
    
    nn.AdaptiveAvgPool2d(1),
    nn.Flatten(),
    nn.Linear(256,512),
    nn.ReLU(),
    nn.Linear(512,5)
)
```

# ⚙ Training Configuration

Model quyidagi parametrlar bilan train qilinadi:

Device:

CUDA mavjud bo‘lsa GPU ishlatiladi
aks holda CPU ishlatiladi

Optimizer:

Adam optimizer ishlatilgan

Learning Rate:

0.001

Loss Function:

CrossEntropyLoss → Multi-class classification uchun

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)
optimizer = torch.optim.Adam(model.parameters(),lr=0.001)
criterion = nn.CrossEntropyLoss()
```

# 🚀 Model Training Loop

Training jarayoni:

Har epoch davomida:

1. Forward pass
2. Loss hisoblash
3. Backpropagation
4. Weight update

Shuningdek:

Training accuracy hisoblab boriladi
Training loss monitoring qilinadi

```python
    for epoch in range(50):
        model.train()
        running_loss = 0.0
        correct = 0
        total = 0

        for inputs, labels in tr_load:
            inputs, labels = inputs.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

        epoch_loss = running_loss / len(tr_load)
        epoch_acc = 100 * correct / total
        print(f"Epoch [{(epoch+1):02d}/{50}] | Loss: {epoch_loss:.4f} | Accuracy: {epoch_acc:.2f}%")
```

# 🧪 Test Accuracy Evaluation

Model test datasetda baholanadi:

torch.no_grad() ishlatiladi:

Bu gradient hisoblashni o‘chiradi va inference tezligini oshiradi

Natijada:

Test accuracy (%) hisoblanadi

```python
model.eval()
total = 0
correct = 0
with torch.no_grad():
    for image,label in te_load:
        image = image.to(device)
        label = label.to(device)
        pr = model(image)
        _, predicted = torch.max(pr.data, 1)
        total += label.size(0)
        correct += (predicted == label).sum().item()
print(f"Accuracy:{((correct / total)*100):.2f}%")
```

# 📊 Validation Accuracy Evaluation

Validation dataset orqali model performance tekshiriladi

Validation:

Training vaqtida model tuning qilish uchun ishlatiladi

Bu orqali:

Overfitting aniqlanadi

```python
total = 0
correct = 0
with torch.no_grad():
    for image,label in vl_load:
        image = image.to(device)
        label = label.to(device)
        pr = model(image)
        _, predicted = torch.max(pr.data, 1)
        total += label.size(0)
        correct += (predicted == label).sum().item()
print(f"Accuracy:{((correct / total)*100):.2f}%")
```

# 🔍 Random Image Prediction Visualization

Model tasodifiy rasmlarda prediction qiladi

Natijalar:

Original image ko‘rsatiladi
Model prediction label chiqariladi

Bu modelning real ishlashini vizual tekshirish imkonini beradi

```python
import matplotlib.pyplot as plt
import numpy as np

def predict_random_images(num_images=4):
    model.eval()
    images, labels = next(iter(vl_load))
    images, labels = images.to(device), labels.to(device)

    outputs = model(images)
    _, preds = torch.max(outputs, 1)

    plt.figure(figsize=(15, 5))
    for i in range(num_images):
        plt.subplot(1, num_images, i+1)
        img = images[i].cpu().numpy().transpose((1, 2, 0))
        img = np.clip(img * [0.229, 0.224, 0.225] + [0.485, 0.456, 0.406], 0, 1)
        plt.imshow(img)
        color = "green" if preds[i] == labels[i] else "red"
        plt.title(f"Real: {classes[labels[i]]} Pred: {classes[preds[i]]}", color=color)
        plt.axis('off')
    plt.show()

predict_random_images()
```
